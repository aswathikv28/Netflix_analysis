```python
import pandas as pd

# Read your data
data = pd.read_csv(r"C:\Users\AJAY\Downloads\archive (1)\netflix_titles.csv")

# Strip leading/trailing spaces from the date strings
data['date_added'] = data['date_added'].str.strip()

# Convert date strings to datetime using the correct format
data['Date_N'] = pd.to_datetime(data['date_added'], errors='coerce', format='%B %d, %Y')

# Check for any dates that couldn't be parsed
invalid_dates = data[data['Date_N'].isnull()]

# Print the rows with invalid dates
print(invalid_dates)

# Print the dataframe to verify successful conversion
print(data.head())
```
